```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Update the file path to the correct location
file_path = "/Users/rajendraacharya/Desktop/Python/AI:ML/training_data.csv"

# Read CSV with a colon (:) separator
train_data = pd.read_csv(file_path, delimiter=":")

# Print the DataFrame
print(train_data)

# Handle null values by filling with mean
train_data.fillna(train_data.mean(), inplace=True)

# Extract features (X) and target variable (y)
X = train_data[['x']]  # Replace 'Feature_Column' with the actual column name
y = train_data['y']  # Replace 'Target_Column' with the actual column name

# Create a linear regression model
model = LinearRegression()

# Fit the model
model.fit(X, y)

# Make predictions
predictions = model.predict(X)

# Plot the data and the regression line
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, predictions, color='red', linewidth=2, label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression Analysis')
plt.legend()
plt.show()

# Calculate Mean Squared Error
mse = mean_squared_error(y, predictions)
print(f"Mean Squared Error: {mse}")
```
